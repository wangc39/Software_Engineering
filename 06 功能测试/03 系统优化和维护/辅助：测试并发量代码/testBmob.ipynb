```python
import threading
from bmob import *

b = Bmob("2731d0d62e86bf7a0d624b9cbda7ad2c","90830d53d6ec9946ba93320a69fde147")

response = b.find("Person")
objectIdList = [item["objectId"] for item in response.jsonData['results']] #objectid是每个人的唯一标识，获取所有obj列表
```

```python
"""
通过实例化threading.Thread类创建线程
"""
import time
import threading

from random import choice

def test_thread(para='hi', sleep=3):
    """线程运行函数"""
    time.sleep(sleep)
    print(para)

def queryDataset():
    
    obj = choice(objectIdList)
    response = b.find("Person",objectId=obj)
    user = response.jsonData #user是一个字典类型，包括数据库中定义的各种属性
    
    print(user['username'],"正在访问")
    time.sleep(0.1)

def main():
    # 创建线程
    thread_list = []
    for i in range(500):
        t = threading.Thread(target=queryDataset)
        thread_list.append(t)
    # 启动线程
    for i in range(500):
        thread_list[i].start()
    
    # print("成功访问率: ", times / 500)


if __name__ == '__main__':
    main()
```
